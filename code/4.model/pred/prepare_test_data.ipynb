```python
sampled_adata = sample[sample.obs.cell_type == 'BP'].obs.sample(n=5, random_state=42)
sample[sampled_adata.index].obs
```

```python
# get indices of 5 samples per cell type
unique_types = np.unique(sample.obs['cell_type'])

sample_obs = []

for t in unique_types:
    sampled_rows = sample[sample.obs.cell_type == t].obs.sample(n=5, random_state=42)
    sample_obs.append(sampled_rows.index)    
```

```python
sample_obs
```

```python
# get rows in dataset corresponding to sampled indices
private_test_input_sample = sample[[elem for index_obj in sample_obs for elem in index_obj.tolist()]]
```
